```python
import pandas as pd
import numpy as np
```

```python
#открываем файл
data = pd.read_csv( 'Mountains.csv', delimiter = ',' )
```

```python
#группируем файл по столбцу 'Parent mountain' и считаем количество упоминаний каждого горного хребта
data1 = data.groupby(['Parent mountain']).size().reset_index(name = 'number of parent mountain')
data1.head(10)
```

```python
#создаем маску и сохраняем полученную таблицу
mask = ( data1['number of parent mountain'] > 3)
data2=data1.loc[ mask ]
data2
```

```python
data2.to_csv('mountain size.csv',index = False)
```

```python
#соединяем исходную таблицу и полученную таблицы
data3 = data.merge( data2, how = 'inner', left_on = 'Parent mountain', right_on = 'Parent mountain' )
data3
```

```python
#сохраняем таблицу
data3.to_csv('mountain size3.csv',index = False)
```

```python
data4 = pd.pivot_table( data3, index = ['Parent mountain', 'number of parent mountain','Mountain', 'Height (m)',
                                        'Height (ft)', 'Prominence (m)' ,'Range', 'Coordinates', 'First ascent', 
                                        'Failed attempts bef. 2004'] )
data4
```
